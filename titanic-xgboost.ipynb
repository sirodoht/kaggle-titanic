```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("./data/train.csv")
```

```python
df
```

```python
df.groupby(["Pclass", "Survived"]).size()
```

```python
df[df["Survived"] == 1]
```

```python
df["SexCategory"] = df["Sex"].map({'male': 1, 'female': -1})
age_mean = df['Age'].mean()
df["Age"].fillna(age_mean, inplace=True)
X_train = df[["PassengerId", "Pclass", "SexCategory", "Age"]]

print(X_train.dtypes)
X_train
```

```python
y_train = df[["Survived"]]
y_train
```

```python
test_data = df[["PassengerId", "Pclass", "SexCategory", "Age"]]
test_data
```

```python
from xgboost import XGBClassifier
```

```python
bst = XGBClassifier(n_estimators=2, max_depth=2, learning_rate=1, objective='binary:logistic')
```

```python
bst.fit(X_train, y_train)
```

```python
df_test = pd.read_csv("./data/test.csv")
df_test["SexCategory"] = df_test["Sex"].map({'male': 1, 'female': -1})
age_mean = df['Age'].mean()
df_test["Age"].fillna(age_mean, inplace=True)
X_test = df_test[["PassengerId", "Pclass", "SexCategory", "Age"]]
X_test
```

```python
preds = bst.predict(X_test)
```

```python
preds
```

```python
len(preds)
```

```python
import csv

with open('preds-xgboost.csv', 'w', newline='') as csvfile:
    fieldnames = ['PassengerId', 'Survived']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for index, item in enumerate(preds):
        writer.writerow({'PassengerId': X_test.loc[index, "PassengerId"], 'Survived': item})
```
